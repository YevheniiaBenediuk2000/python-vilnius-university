1. Create a class "BankAccount". The class constructor has to accept two parameters: "name" - account holder's name and "balance" - account balance. Create a class method "add_transaction" which accept amount of money to be debited / added to the current account balance. Create a "current_balance" method to return the current account balance. 

```python
class BankAccount: 
    def __init__(self, name, balance):
        self.name = name
        self.balance = balance

    def add_transaction(self, amount_debited):
        self.balance += amount_debited
        
    def current_balance(self):
        print(f"Current balance of {self.name} is: {self.balance}")
        return self.balance

uk_prime_minister_account = BankAccount("Boris Johnson", 1000000)
princess_diana_account = BankAccount("Diana Frances Spencer", 1000000000)

uk_prime_minister_account.add_transaction(1000)
princess_diana_account.add_transaction(9999)

uk_prime_minister_account.current_balance()
princess_diana_account.current_balance()
```

2. Supplement the first task class with the ability to store all transactions. Also create a "show_transactions" method that prints all the transactions on the account. 

```python
class BankAccount: 
    def __init__(self, name, balance):
        self.name = name
        self.balance = balance
        self.transactions = []

    def add_transaction(self, amount_debited):
        self.balance += amount_debited
        self.transactions.append(amount_debited)
        
    def show_transactions(self):
        print(f"The list of transactions for {self.name}:")
        for transaction in self.transactions:
            if transaction > 0:
                print(f"+{transaction}")
            else:
                print(f"{transaction}")
        
    def current_balance(self):
        print(f"Current balance of {self.name} is: {self.balance}")
        return self.balance
    
uk_prime_minister_account = BankAccount("Boris Johnson", 1000000)
princess_diana_account = BankAccount("Diana Frances Spencer", 1000000000)

uk_prime_minister_account.add_transaction(1000)
uk_prime_minister_account.add_transaction(2000)
princess_diana_account.add_transaction(9999)

uk_prime_minister_account.current_balance()
princess_diana_account.current_balance()

uk_prime_minister_account.show_transactions()
princess_diana_account.show_transactions()
```

3. Supplement the first task class with the static method "is_balance_positive", which accept current balance and would return True if the balance is positive and False if negative. 

```python
class BankAccount: 
    def __init__(self, name, balance):
        self.name = name
        self.balance = balance
        self.transactions = []

    def add_transaction(self, amount_debited):
        self.balance += amount_debited
        self.transactions.append(amount_debited)
        
    def show_transactions(self):
        print(f"The list of transactions for {self.name}:")
        for transaction in self.transactions:
            if transaction > 0:
                print(f"+{transaction}")
            else:
                print(f"{transaction}")
        
    def current_balance(self):
        print(f"Current balance of {self.name} is: {self.balance}")
        return self.balance

    @staticmethod
    def is_balance_positive(balance):
        if balance > 0:
            print(f"Balance is positive")
            return True
        else:
            print(f"Balance is negative")
            return False
    
uk_prime_minister_account = BankAccount("Boris Johnson", 1000000)
princess_diana_account = BankAccount("Diana Frances Spencer", 1000000000)

uk_prime_minister_account.add_transaction(1000)
uk_prime_minister_account.add_transaction(2000)
princess_diana_account.add_transaction(9999)

uk_prime_minister_account.current_balance()
princess_diana_account.current_balance()

uk_prime_minister_account.show_transactions()
princess_diana_account.show_transactions()

uk_prime_minister_account.is_balance_positive(uk_prime_minister_account.balance)
princess_diana_account.is_balance_positive(princess_diana_account.balance)
```

4. Supplement the first task class with the class method "create_multiple", which will accept list with names and surnames, and would return a list of accounts with zero balances.

```python
class BankAccount: 
    def __init__(self, name, balance):
        self.name = name
        self.balance = balance
        self.transactions = []

    def add_transaction(self, amount_debited):
        self.balance += amount_debited
        self.transactions.append(amount_debited)
        
    def show_transactions(self):
        print(f"The list of transactions for {self.name}:")
        for transaction in self.transactions:
            if transaction > 0:
                print(f"+{transaction}")
            else:
                print(f"{transaction}")
        
    def current_balance(self):
        print(f"Current balance of {self.name} is: {self.balance}")
        return self.balance

    @staticmethod
    def is_balance_positive(balance):
        if balance > 0:
            print(f"Balance is positive")
            return True
        else:
            print(f"Balance is negative")
            return False
        
    @classmethod
    def create_multiple(cls, names_surnames):
        accounts = []

        for name_surname in names_surnames:
            account = cls(name_surname, 0)
            accounts.append(account)
        return accounts
    
uk_prime_minister_account = BankAccount("Boris Johnson", 1000000)
princess_diana_account = BankAccount("Diana Frances Spencer", 1000000000)

uk_prime_minister_account.add_transaction(1000)
uk_prime_minister_account.add_transaction(2000)
princess_diana_account.add_transaction(9999)

uk_prime_minister_account.current_balance()
princess_diana_account.current_balance()

uk_prime_minister_account.show_transactions()
princess_diana_account.show_transactions()

uk_prime_minister_account.is_balance_positive(uk_prime_minister_account.balance)
princess_diana_account.is_balance_positive(princess_diana_account.balance)

new_accounts = BankAccount.create_multiple(["John Doe", "Jane Smith", "Alice Johnson"])

for account in new_accounts:
    print(f"Account for {account.name} created with balance {account.balance}")
```

5. Create an Address Book application. Use two classes: 
    * "Contact" - which describes the contact made up of attributes: first name, last name, phone number, email address, workplace.
    * "AddressBook" - where all your contacts are stored. 
    
   The following functions must be implemented: creating, editing, deleting contacts, searching for a contact by name, surname and starting the phone number. A method is needed to derive the number of contacts. It also requires a nice display of the contact and address book on the screen, with the ability to sort by name or surname. 
   
   Main menu example:
   1. Create a contact (enter name, surname, phone number, email, workplace)
   2. Edit a contact (enter contact for editing, enter new name, surname, phone number, email, workplace)
   3. Remove a contact (enter contact for removing)
   4. Number of contacts in AddressBook
   5. Search contact by:
       1. Name
       2. Surname
       3. Phone number
   6. Whole AddressBook
       1. Sort by name
       2. Sort by surname

```python
class Contact:
    def __init__(self, first_name, last_name, phone_number, email, workplace):
        self.first_name = first_name
        self.last_name = last_name
        self.phone_number = phone_number
        self.email = email
        self.workplace = workplace

    def __str__(self):
        return f"Contact(Name: {self.first_name}, Last Name: {self.last_name}, Phone number: {self.phone_number}, Email: {self.email}, Workplace: {self.workplace})"

class AddressBook():
    def __init__(self):
        self.contacts = []

    def create_contact(self):
        print("Enter the contact details:")
        first_name = input("Name: ")
        last_name = input("Surname: ")
        phone_number = input("Phone number: ")
        email = input("Email address: ")
        workplace = input("Workplace: ")

        new_contact = Contact(first_name, last_name, phone_number, email, workplace)
        self.contacts.append(new_contact)
        print("Contact added successfully! ")
        print(new_contact)

    def print_contacts(self):
        if not self.contacts:
            print("No contacts to display")
            return 
         
        print("Available contacts: ")
        for index, contact in enumerate(self.contacts):
            print(f"{index+1}: {contact.first_name}, {contact.last_name}, {contact.phone_number}, {contact.email}, {contact.workplace}")

    def edit_contact(self):
        self.print_contacts()

        try:
            choice = int(input("Select the number of the contact to edit: ")) - 1
            selected_contact = self.contacts[choice]

            selected_contact.first_name = input(f"Enter the first name (current: {selected_contact.first_name}) ")
            selected_contact.last_name = input(f"Enter the last name (current: {selected_contact.last_name}) ")
            selected_contact.phone_number = input(f"Enter the phone number (current: {selected_contact.phone_number}) ")
            selected_contact.email = input(f"Enter the email (current: {selected_contact.email}) ")
            selected_contact.workplace = input(f"Enter the workplace (current: {selected_contact.workplace}) ")
            
        except ValueError:
            print("Invalid contact number")
        except IndexError:
            print("Invalid input. Please enter a valid number")

    def remove_contact(self):
        self.print_contacts()
        try:
            choice = int(input("Select the number of the contact to remove: ")) - 1
            if choice < 0 or choice >= len(self.contacts):
                print("Invalid input. Please enter a valid number")
                return 
            
            del self.contacts[choice]
            print("Contact removed successfully.")
        except ValueError:
            print("Invalid input. Enter a valid number.")

    def number_of_contacts(self):
        if not self.contacts:
            print("No contacts to display")
            return 
            
        print(f"The number of contacts are: {len(self.contacts)}")
            
    def contact_search(self, search_criteria):
        found_contact = []

        # Search by name
        if search_criteria == 1:
            search_name = input("Enter the name to search for: ").lower()
            for contact in self.contacts:
                if search_name in contact.first_name.lower():
                    found_contact.append(contact)

        # Search by surname
        elif search_criteria == 2:
            search_surname = input("Enter the surname to search for: ").lower()
            for contact in self.contacts:
                if search_surname in contact.last_name.lower():
                    found_contact.append(contact)

        # Search by phone number
        elif search_criteria == 3:
            search_phone_number = input("Enter the phone number to search for: ").lower()            
            for contact in self.contacts:
                if search_phone_number in contact.phone_number.lower():
                    found_contact.append(contact)
        else:
            print("Invalid search criteria")

        # Print the found contacts
        if not found_contact:
            print("No contacts found.")
        else:
            print("Found contacts: ")
            for contact in found_contact:
                print(f"{contact.first_name} {contact.last_name} {contact.phone_number} {contact.email} {contact.workplace}")

    def contact_sort(self, sort_criteria):
        # Search by name
        if sort_criteria == 1:
            sorted_contacts = sorted(self.contacts, key=lambda contact: contact.first_name.lower())
        elif sort_criteria == 2:
            sorted_contacts = sorted(self.contacts, key=lambda contact: contact.last_name.lower())
        else:
            print("Invalid sort criteria")
            return
        for contact in sorted_contacts:
            print(contact)

address_book = AddressBook()

while True:
    print ("\nMain Menu, please choose an option:")
    print("1. Create a contact")
    print("2. Edit a contact")
    print("3. Remove a contact")
    print("4. Number of contacts in AddressBook")
    print("5. Search contact")
    print("6. Whole AddressBook")
    try:
        choice = int(input("Enter your choice: "))
    except ValueError:
        print("Invalid input, please enter a number")
        continue

    if choice == 1:
        address_book.create_contact()
    elif choice == 2:
        address_book.edit_contact()
    elif choice == 3:
        address_book.remove_contact()
    elif choice == 4:
        address_book.number_of_contacts()
    elif choice == 5:
        print("1. Name")
        print("2. Surname")
        print("3. Phone number")
        search_criteria = int(input("Enter your choice: "))
        address_book.contact_search(search_criteria)
    elif choice == 6:
        print("1. Sort by name")
        print("2. Sort by surname")
        sort_criteria = int(input("Enter your choice: "))
        address_book.contact_sort(sort_criteria)
```
